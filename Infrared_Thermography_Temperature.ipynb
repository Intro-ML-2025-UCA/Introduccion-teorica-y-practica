```python
# ==================================================
# 2. IMPORTAR LIBRERÍAS
# ==================================================
from ucimlrepo import fetch_ucirepo
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# ==================================================
# 3. DESCARGAR EL DATASET
# ==================================================
print("Descargando dataset...")
infrared_thermography_temperature = fetch_ucirepo(id=925)

# Separar características (X) y objetivo (y)
X = infrared_thermography_temperature.data.features
y = infrared_thermography_temperature.data.targets

print("✓ Dataset descargado exitosamente\n")
```

```python

# ==================================================
# 4. PRIMERA EXPLORACIÓN
# ==================================================
print("=" * 60)
print("PRIMERAS 5 FILAS DE CARACTERÍSTICAS (X)")
print("=" * 60)
print(X.head())
print("\n")

print("=" * 60)
print("PRIMERAS 5 FILAS DEL OBJETIVO (y)")
print("=" * 60)
print(y.head())
print("\n")

```

```python
# ==================================================
# 5. NOMBRES DE LAS VARIABLES
# ==================================================
print("=" * 60)
print("NOMBRES DE LAS COLUMNAS")
print("=" * 60)
print("Características disponibles en X:")
print(X.columns.tolist())
print(f"\nTotal de características: {len(X.columns)}")
print("\nVariable objetivo en y:")
print(y.columns.tolist())
print("\n")
```

```python
# ==================================================
# 6. ESTADÍSTICAS DESCRIPTIVAS
# ==================================================
print("=" * 60)
print("ESTADÍSTICAS DESCRIPTIVAS DE X")
print("=" * 60)
print(X.describe())
print("\n")

print("=" * 60)
print("ESTADÍSTICAS DESCRIPTIVAS DE y (Temperatura Oral)")
print("=" * 60)
print(y.describe())
print("\n")
```

```python
# ==================================================
# 7. VISUALIZACIONES
# ==================================================

# Gráfico 1: Dispersión Edad vs Temperatura
if 'Age' in X.columns:
    plt.figure(figsize=(10, 6))
    plt.scatter(X['Age'], y[y.columns[0]], alpha=0.5, c='blue', edgecolors='black')
    plt.xlabel('Edad', fontsize=12)
    plt.ylabel('Temperatura Oral (°F)', fontsize=12)
    plt.title('Relación: Edad vs Temperatura Oral', fontsize=14, fontweight='bold')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

# Gráfico 2: Histograma de Temperatura Oral
plt.figure(figsize=(10, 6))
plt.hist(y[y.columns[0]], bins=30, edgecolor='black', alpha=0.7, color='coral')
plt.xlabel('Temperatura Oral (°F)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.title('Distribución de Temperaturas Orales', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
```
